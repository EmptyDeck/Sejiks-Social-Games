```python
import json

# 텍스트 파일에서 JSON 데이터 읽기
with open("jack.txt", "r", encoding="utf-8") as file:
    data = json.load(file)

# 사람이 읽기 쉽게 출력
for entry in data["content"]:
    print(f"Category: {entry['category']}")
    print(" Tasks:")
    for task in entry["tasks"]:
        print(f"  - {task['v']}")
    print("-" * 40)
```
